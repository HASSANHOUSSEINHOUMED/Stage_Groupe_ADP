```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

urls = [
    'https://shop.frankfurt-airport.com/en-US/product/boss-orange-woman-eau-de-toilette-30-ml_dcdf59254787e913',
    'https://shop.frankfurt-airport.com/en-US/product/issey-miyake-fusion-d-issey-extreme-eau-de-toilett_43cda6189e5ab5bf',
    'https://shop.frankfurt-airport.com/en-US/product/prada-infusion-figue-eau-de-parfum-100-ml_8b7ee55f1d85b0f1',
    'https://shop.frankfurt-airport.com/en-US/product/ralph-lauren-polo-eau-de-toilette-118-ml_49c6d41c8a5c55fa'
]

data = []

for url in urls:
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        product_data = {}
        product_data['URL'] = url

        product_data['Product Order Amount'] = soup.find('dd', {'data-testid': 'productAttribute-orderAmount'}).text.strip() if soup.find('dd', {'data-testid': 'productAttribute-orderAmount'}) else None
        product_data['Brand'] = soup.find('a', class_='ProductHeader_link__4G3zn').text.strip() if soup.find('a', class_='ProductHeader_link__4G3zn') else None
        product_data['Name'] = soup.find('h1', class_='Headline_headline___0NwR Headline_h1__8kl7_').text.strip() if soup.find('h1', class_='Headline_headline___0NwR Headline_h1__8kl7_') else None
        product_data['Price'] = soup.find('div', class_='ProductPriceRaw_actualPrice__U7PIE').text.strip() if soup.find('div', class_='ProductPriceRaw_actualPrice__U7PIE') else None
        product_data['Description'] = soup.find('div', class_='ProductDescription_richtext__pOzsV Richtext_richText__OQl_C').text.strip() if soup.find('div', class_='ProductDescription_richtext__pOzsV Richtext_richText__OQl_C') else None
        product_data['Gender'] = soup.find('div', class_='ProductAttributeDefaultFormatter_text__YAOZX Richtext_richText__OQl_C').text.strip() if soup.find('div', class_='ProductAttributeDefaultFormatter_text__YAOZX Richtext_richText__OQl_C') else None

        made_in_dd = soup.find('dd', {'data-testid': 'productAttribute-20000743_countryZoneOfOrigin'})
        product_data['Made In'] = made_in_dd.find('div', class_='ProductAttributeDefaultFormatter_text__YAOZX').text.strip() if made_in_dd else None

        fragrance_dd = soup.find('dd', {'data-testid': 'productAttribute-20000298_typeOfFragrance'})
        product_data['Fragrance Type'] = fragrance_dd.find('div', class_='ProductAttributeDefaultFormatter_text__YAOZX').text.strip() if fragrance_dd else None

        warning_dd = soup.find('dd', {'data-testid': 'productAttribute-warnings'})
        product_data['Warning Hints'] = warning_dd.find('div', class_='ProductAttributeDefaultFormatter_text__YAOZX').text.strip() if warning_dd else None

        item_number_dd = soup.find('dd', {'data-testid': 'productAttribute-sku'})
        product_data['Item Number'] = item_number_dd.find('div', class_='ProductAttributeDefaultFormatter_text__YAOZX Richtext_richText__OQl_C').text.strip() if item_number_dd else None

        product_series_dd = soup.find('dd', {'data-testid': 'productAttribute-manufacturerSeriesname'})
        product_data['Series'] = product_series_dd.find('div', class_='ProductAttributeDefaultFormatter_text__YAOZX Richtext_richText__OQl_C').text.strip() if product_series_dd else None

        product_data['Weight'] = soup.find('dd', {'data-testid': 'productAttribute-netWeight'}).text.strip() if soup.find('dd', {'data-testid': 'productAttribute-netWeight'}) else None
        product_data['Height'] = soup.find('dd', {'data-testid': 'productAttribute-packageHeight'}).text.strip() if soup.find('dd', {'data-testid': 'productAttribute-packageHeight'}) else None
        product_data['Width'] = soup.find('dd', {'data-testid': 'productAttribute-packageWidth'}).text.strip() if soup.find('dd', {'data-testid': 'productAttribute-packageWidth'}) else None
        product_data['Package Length'] = soup.find('dd', {'data-testid': 'productAttribute-packageLength'}).text.strip() if soup.find('dd', {'data-testid': 'productAttribute-packageLength'}) else None

        data.append(product_data)
    else:
        print(f'Échec de la requête HTTP pour {url}, status code:', response.status_code)

# Créez un DataFrame pandas à partir des données collectées
df = pd.DataFrame(data)

# Affichez le DataFrame
print(df)

# Optionnel: Enregistrer le DataFrame dans un fichier CSV
df.to_csv('product_data_first_website.csv', index=False)
```

# Scrapping dubai duty free en Python

```python
# Installer Selenium avec la vérification SSL désactivée
# !pip install selenium --trusted-host pypi.org --trusted-host files.pythonhosted.org

# Installer WebDriver Manager avec la vérification SSL désactivée
# !pip install webdriver-manager --trusted-host pypi.org --trusted-host files.pythonhosted.org

import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO

# Chemin chromedriver.exe
chromedriver_path = r'C:\Users\HOUSSEIH\AppData\Local\binman\binman_chromedriver\win32\125.0.6422.78\chromedriver.exe'

# Configurer le navigateur Chrome
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--headless")  # Optionnel: exécuter Chrome en mode headless (sans interface graphique)
chrome_options.add_argument("--window-size=1920,1080")

# Initialiser le service Chrome avec le chemin spécifié
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Naviguer vers la page principale pour accepter les cookies une seule fois
url_main = "https://www.dubaidutyfree.com/"
driver.get(url_main)
time.sleep(5)  # Attendre que la page charge complètement

# Gérer les cookies une seule fois
try:
    print("Tentative de recherche du bouton 'Accepter les cookies'")
    accept_cookies_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Accept')]"))
    )
    print("Bouton 'Accepter les cookies' trouvé. Tentative de clic.")
    accept_cookies_button.click()
    time.sleep(5)
    print("Clic sur le bouton 'Accepter les cookies' réussi.")
except Exception as e:
    print(f"Erreur lors de la tentative de clic sur le bouton 'Accepter les cookies': {e}")

# Liste des produits
list_des_produits = [
    "byredo-gypsy-water-roll-on-perfume-oil-75ml/product/111484141",
    "dior-sauvage-parfum-100ml/product/111413943",
    "giorgio-armani-code-homme-le-parfum-75ml/product/113603240",
    "hugo-boss-alive-parfum-50ml/product/114098990"
]

# DataFrame pour stocker les spécifications
specifications = pd.DataFrame()

# Boucle pour scraper les informations de chaque produit
for product in list_des_produits:
    url = f"https://www.dubaidutyfree.com/{product}"
    driver.get(url)
    time.sleep(10)  # Attendre que la page charge complètement

    # Extraire le HTML de la page et le traiter avec BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Extraire les spécifications
    tables = soup.find_all('table', class_='cameraSpecification')
    for table in tables:
        # Utiliser StringIO pour encapsuler la chaîne de caractères HTML
        df = pd.read_html(StringIO(str(table)))[0].transpose()
        df.columns = df.iloc[0]
        df = df[1:]
        specifications = pd.concat([specifications, df], ignore_index=True)

    # Extraire et traiter le prix
    try:
        price_text = soup.find('span', class_='listPriceValue').text.replace(',', '').split('.')[0]
        specifications['Price'] = float(price_text) / 4.03  # Conversion en euros supposée
    except AttributeError:
        print(f"Prix non trouvé pour {product}")
        specifications['Price'] = None

# Fermer le navigateur
driver.quit()

# Afficher le DataFrame
print(specifications)

# Enregistrer le DataFrame dans un fichier CSV
specifications.to_csv(r'C:\Users\HOUSSEIH\Documents\Scraping\Specifications Python.csv', index=False)
```
