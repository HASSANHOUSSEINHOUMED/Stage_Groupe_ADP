```python
import pandas as pd
from langdetect import detect
from deep_translator import GoogleTranslator
from tqdm.auto import tqdm

def lang_detect(x):
    """Detect the language of a review using the langdetect package."""
    try:
        return detect(x)
    except:
        return None  # Safely handle exception if detection fails

def translate(x, translator, target_lang='en'):
    """Translate a review using the specified translation tool."""
    try:
        return translator.translate(x, target_lang=target_lang) if x else x
    except:
        return x  # Return the original text if translation fails

def apply_translation_and_detection(df, source_col, target_col, lang_col, target_lang='en'):
    """Translate text from a specified source column to a target column in the DataFrame and detect language."""
    translator = GoogleTranslator(source='auto', target=target_lang)

    # Language detection with progress bar
    print("Starting language detection...")
    df[lang_col] = [lang_detect(text) for text in tqdm(df[source_col], desc="Detecting languages")]

    # Translation with progress bar
    print("Starting translation...")
    df[target_col] = [translate(text, translator, target_lang) if lang != target_lang else text
                      for text, lang in tqdm(zip(df[source_col], df[lang_col]), total=len(df), desc="Translating reviews")]
    return df
```

## Performances en Clustering :
Ce modèle a montré de très bonnes performances dans des tâches de clustering de textes, ce qui est essentiel pour la modélisation de sujets où les documents similaires doivent être groupés ensemble efficacement.
## Généralisation :
Il a été entraîné pour capturer des relations sémantiques complexes à travers différents domaines de texte, ce qui le rend efficace pour une variété d'applications, y compris la modélisation de sujets sur des corpus multilingues ou spécifiques à un domaine.

# SetFit One Label by review Classification Avec les bonnes hyperparamètres

### Voici les principaux hyperparamètres que j'envisage d'optimiser :

#### 1. Taux d'Apprentissage (Learning Rate)
Importance: Le taux d'apprentissage contrôle la vitesse à laquelle le modèle apprend. Un taux trop élevé peut conduire à un apprentissage instable, tandis qu'un taux trop faible peut rendre l'apprentissage trop lent, avec le risque de rester bloqué dans des minimums locaux.
Impact: Ajuster le taux d'apprentissage peut aider à trouver le meilleur compromis entre rapidité de convergence et stabilité de l'apprentissage.
#### 2. Nombre d'Époques (Num Train Epochs)
Importance: Le nombre d'époques détermine combien de fois le modèle va voir l'ensemble des données d'entraînement. Trop peu d'époques peuvent conduire à un sous-apprentissage, tandis que trop d'époques peuvent causer un surapprentissage.
Impact: Optimiser le nombre d'époques peut aider à maximiser la performance du modèle sur les données de validation ou de test sans overfitting.
#### 3. Taille du Batch (Per Device Train Batch Size)
Importance: La taille du batch influence la stabilité de l'estimation du gradient lors de l'entraînement. Une taille de batch plus grande offre une estimation plus stable, mais à un coût computationnel plus élevé et parfois une qualité de généralisation moindre.
Impact: Trouver la bonne taille de batch peut améliorer l'efficacité et l'efficience de l'entraînement.
#### 4. Poids de la Régularisation (Weight Decay)
Importance: Le weight decay est une forme de régularisation L2 qui aide à prévenir le surapprentissage en ajoutant une pénalité sur la taille des poids du modèle.
Impact: Ajuster le poids de la régularisation peut aider à garder le modèle généralisable à de nouvelles données, en évitant qu'il ne s'adapte trop finement aux données d'entraînement.

### Pourquoi Optimiser Ces Hyperparamètres?
Optimiser ces hyperparamètres dans mon contexte spécifique de classification de texte peut apporter plusieurs avantages :

#### Amélioration de la Précision:
En ajustant ces paramètres, vous pouvez améliorer la capacité du modèle à classifier correctement les nouvelles entrées, augmentant ainsi la précision globale.
#### Éviter le Surapprentissage et le Sous-apprentissage:
Une bonne combinaison de paramètres aide à équilibrer entre l'apprentissage suffisant des caractéristiques des données et le maintien de la capacité du modèle à bien généraliser.
#### Efficacité Computationnelle:
Optimiser par exemple la taille du batch et le taux d'apprentissage peut également affecter la vitesse d'entraînement du modèle, ce qui est crucial lorsque vous travaillez avec de grandes quantités de données ou des ressources de calcul limitées.


# Optuna

Optuna nécessite une seule métrique comme valeur de retour pour comparer les performances des différentes configurations d'hyperparamètres.

```python
# Etape 1 : recherche des bonnes hyperparamètres

import numpy as np
import pandas as pd
import optuna
import mlflow
import torch
from datasets import load_metric, Dataset, DatasetDict
from sklearn.preprocessing import LabelEncoder
from transformers import AutoModelForSequenceClassification, AutoTokenizer, TrainingArguments, Trainer

# Initialisation des métriques d'évaluation
accuracy_metric = load_metric("accuracy")
precision_metric = load_metric("precision")
recall_metric = load_metric("recall")
f1_metric = load_metric("f1")

# Initialiser le tokenizer et le nom du modèle
model_name = "sentence-transformers/paraphrase-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Fonction pour tokenizer les données
def tokenize_function(examples):
    return tokenizer(examples['translated_content'], padding='max_length', truncation=True, max_length=4)

# Fonction pour calculer les métriques à partir des prédictions
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy_metric.compute(predictions=predictions, references=labels)["accuracy"],
        "precision": precision_metric.compute(predictions=predictions, references=labels)["precision"],
        "recall": recall_metric.compute(predictions=predictions, references=labels)["recall"],
        "f1": f1_metric.compute(predictions=predictions, references=labels)["f1"]
    }

# Fonction objectif pour l'optimisation avec Optuna et MLflow
def objective(trial):
    # Configuration de MLflow
    mlflow.set_tracking_uri('/content/drive/MyDrive/Colab Notebooks/mlruns_setfit_onelabel')
    mlflow.set_experiment("Setfit Onelabel Experiment")
    mlflow.start_run()

    # Chargement des données et préparation
    df_setfit_onelabel = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/df_setfit_onelabel.csv')
    label_encoder = LabelEncoder()
    df_setfit_onelabel['labels'] = label_encoder.fit_transform(df_setfit_onelabel['labels'])
    dataset = Dataset.from_pandas(df_setfit_onelabel[['translated_content', 'labels']])
    tokenized_datasets = dataset.map(tokenize_function, batched=True)
    train_test_split = tokenized_datasets.train_test_split(test_size=0.1)
    tokenized_datasets = DatasetDict({
        'train': train_test_split['train'],
        'test': train_test_split['test']
    })

    # Suggestion des hyperparamètres
    learning_rate = trial.suggest_loguniform('learning_rate', 1e-5, 1e-3)
    batch_size = trial.suggest_categorical('batch_size', [2, 4, 8])
    num_train_epochs = trial.suggest_int('num_train_epochs', 1, 2)

    # Configuration des arguments de formation
    training_args = TrainingArguments(
        output_dir='/content/drive/MyDrive/Colab Notebooks/results',
        evaluation_strategy="epoch",
        learning_rate=learning_rate,
        per_device_train_batch_size=batch_size,
        num_train_epochs=num_train_epochs,
        weight_decay=0.01,
    )

    # Initialisation du modèle et du trainer
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(label_encoder.classes_))
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=tokenized_datasets['train'],
        eval_dataset=tokenized_datasets['test'],
        tokenizer=tokenizer,
        compute_metrics=compute_metrics
    )

    # Entraînement du modèle
    trainer.train()
    eval_result = trainer.evaluate()

    # Logging des paramètres et des métriques dans MLflow
    mlflow.log_params({
        "learning_rate": learning_rate,
        "batch_size": batch_size,
        "num_train_epochs": num_train_epochs
    })
    mlflow.log_metrics({
        "accuracy": eval_result['eval_accuracy'],
        "precision": eval_result['eval_precision'],
        "recall": eval_result['eval_recall'],
        "f1": eval_result['eval_f1']
    })

    mlflow.end_run()

    return eval_result["eval_accuracy"]

# Exécution de l'étude Optuna
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=2)

print("Meilleurs hyperparamètres : ", study.best_trial.params)
```

```python
# Etape 2 : Re-formation avec les meilleurs hyperparamètres et Sauvegarde du meilleur modèle

df_setfit_onelabel = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/df_setfit_onelabel.csv')
label_encoder = LabelEncoder()
df_setfit_onelabel['labels'] = label_encoder.fit_transform(df_setfit_onelabel['labels'])

dataset = Dataset.from_pandas(df_setfit_onelabel[['translated_content', 'labels']])
tokenized_datasets = dataset.map(tokenize_function, batched=True)
train_test_split = tokenized_datasets.train_test_split(test_size=0.1)
tokenized_datasets = DatasetDict({
        'train': train_test_split['train'],
        'test': train_test_split['test']
    })

# Re-former le modèle avec les meilleurs hyperparamètres
best_params = study.best_trial.params
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(label_encoder.classes_))  # Réinitialisation du modèle

training_args = TrainingArguments(
    output_dir='/content/drive/MyDrive/Colab Notebooks/results',
    evaluation_strategy="epoch",
    learning_rate=best_params['learning_rate'],
    per_device_train_batch_size=best_params['batch_size'],
    num_train_epochs=best_params['num_train_epochs'],
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['test'],
    tokenizer=tokenizer,
)

trainer.train()

# Sauvegarde du modèle entraîné et du tokenizer
output_dir = '/content/drive/MyDrive/Colab Notebooks/setfit_model_onelabel'
model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)
```

```python
# Etape 1 : Optimisation des Hyperparamètres avec Optuna

import numpy as np
import pandas as pd
import optuna
import mlflow
from datasets import load_dataset, load_metric, Dataset, DatasetDict
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from torch import tensor

# Chargement des données
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/df_setfit_onelabel.csv")


# Initialisation du tokenizer et du modèle
MODEL = "yangheng/deberta-v3-base-absa-v1.1"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
# num_labels=3 est pour indiquer trois catégories de sentiments (negative,neutral,positive)
model = AutoModelForSequenceClassification.from_pretrained(MODEL, num_labels=3)


def prepare_dataset(df):
    texts = df['translated_content'].tolist()
    aspects = df['labels_setfit'].tolist()
    label_encoder = LabelEncoder()
    labels = label_encoder.fit_transform(aspects)  # Utiliser les étiquettes encodées pour le dataset
    merged_texts = [f"[CLS] {text} [SEP] {aspect} [SEP]" for text, aspect in zip(texts, aspects)]
    encoded_inputs = tokenizer(merged_texts, padding='max_length', truncation=True, max_length=4, return_tensors="pt")

    return Dataset.from_dict({
        'input_ids': encoded_inputs['input_ids'],
        'attention_mask': encoded_inputs['attention_mask'],
        'labels': tensor(labels)
    })

df_train, df_test = train_test_split(df, test_size=0.1, random_state=42)

# Préparation des datasets
train_dataset = prepare_dataset(df_train)
eval_dataset = prepare_dataset(df_test)


# Métriques pour évaluer le modèle
accuracy = load_metric("accuracy")
precision = load_metric("precision")
recall = load_metric("recall")
f1 = load_metric("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy.compute(predictions=predictions, references=labels)["accuracy"],
        "precision": precision.compute(predictions=predictions, references=labels)["precision"],
        "recall": recall.compute(predictions=predictions, references=labels)["recall"],
        "f1": f1.compute(predictions=predictions, references=labels)["f1"]
    }

# Fonction objectif pour Optuna
def objective(trial):
    # Configuration de MLflow
    mlflow.set_tracking_uri("/content/drive/MyDrive/Colab Notebooks/mlruns_absa")
    mlflow.set_experiment("ABSA Experiment")
    mlflow.start_run()

    learning_rate = trial.suggest_loguniform('learning_rate', 1e-5, 1e-3)
    batch_size = trial.suggest_categorical("batch_size", [2, 4, 8])
    num_epochs = trial.suggest_int("num_epochs", 1, 2)

    training_args = TrainingArguments(
        output_dir="/content/drive/MyDrive/Colab Notebooks/results_absa",
        evaluation_strategy="epoch",
        learning_rate=learning_rate,
        per_device_train_batch_size=batch_size,
        num_train_epochs=num_epochs,
        weight_decay=0.01
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=eval_dataset,
        compute_metrics=compute_metrics,
        tokenizer=tokenizer
    )

    trainer.train()
    eval_result = trainer.evaluate()

    # Enregistrement des résultats dans MLflow
    mlflow.log_params(trial.params)
    mlflow.log_metrics({
        "accuracy": eval_result["eval_accuracy"],
        "precision": eval_result["eval_precision"],
        "recall": eval_result["eval_recall"],
        "f1": eval_result["eval_f1"]
    })

    mlflow.end_run()
    # les classes sont desequilibré donc f1 va m'aider à minimiser à la fois les faux positifs et les faux négatifs, cela permettre d'équilibrer au modèle la precision et le rappel
    return eval_result["eval_f1"]

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=2)

print("Meilleurs hyperparamètres : ", study.best_trial.params)
```

```python
# Etape 2 : Re-formation avec les meilleurs hyperparamètres et Sauvegarde du meilleur modèle

from transformers import TrainingArguments, Trainer

# Charger le modèle avec les meilleurs hyperparamètres
best_params = study.best_trial.params
training_args = TrainingArguments(
    output_dir="/content/drive/MyDrive/Colab Notebooks/results_absa_best",
    evaluation_strategy="epoch",
    learning_rate=best_params['learning_rate'],
    per_device_train_batch_size=best_params['batch_size'],
    num_train_epochs=best_params['num_epochs'],
    weight_decay=0.01
)

# Création d'un nouveau trainer avec les meilleurs paramètres
best_trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    compute_metrics=compute_metrics,
    tokenizer=tokenizer
)

# Entraînement avec les meilleurs paramètres
best_trainer.train()

# Sauvegarde du modèle et du tokenizer entraînés
model.save_pretrained("/content/drive/MyDrive/Colab Notebooks/absa_model")
tokenizer.save_pretrained("/content/drive/MyDrive/Colab Notebooks/absa_model")
```

```python
# Etape 3 : Faire des predictions avec le meilleur modèle

from transformers import AutoModelForSequenceClassification, AutoTokenizer
import torch

input_dir = "/content/drive/MyDrive/Colab Notebooks/absa_model"

# Charger le tokenizer et le modèle depuis le répertoire de sauvegarde
tokenizer = AutoTokenizer.from_pretrained(input_dir)
model = AutoModelForSequenceClassification.from_pretrained(input_dir, num_labels=3)
model.eval

def predict(text, aspect, tokenizer, model):

    # Préparation de l'entrée
    inputs = tokenizer(f"[CLS] {text} [SEP] {aspect} [SEP]", return_tensors="pt", padding=True, truncation=True, max_length=512)

    # Prédiction
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        probabilities = torch.softmax(logits, dim=-1)
        predicted_class = torch.argmax(probabilities, dim=-1)
        predicted_label = ['negative', 'neutral', 'positive'][predicted_class.item()]
        predicted_score = probabilities[0, predicted_class.item()].item()

    return predicted_label, predicted_score

# Exemple de test
text = "The food was great but the service was terrible"
aspect = "service"
predicted_label, predicted_score = predict(text, aspect, tokenizer, model)
print(f"Predicted label: {predicted_label}, Score: {predicted_score}")
```

```python
# Application de la fonction de prédiction à chaque ligne du DataFrame
results = df.apply(lambda row: predict(row['translated_content'], row['labels_setfit'], tokenizer, model), axis=1)

# Création de nouvelles colonnes pour les labels prédits et les scores
df['predicted_label'] = results.apply(lambda x: x[0])
df['predicted_score'] = results.apply(lambda x: x[1])
```

```python
# Etape 1 : recherche des bonnes hyperparamètres

import numpy as np
import pandas as pd
import optuna
import mlflow
import torch
from sklearn.model_selection import train_test_split
from datasets import load_metric, Dataset, DatasetDict
from sklearn.preprocessing import LabelEncoder, MultiLabelBinarizer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from transformers import AutoModelForSequenceClassification, AutoTokenizer, TrainingArguments, Trainer

# Initialisation des métriques d'évaluation
accuracy_metric = load_metric("accuracy")
precision_metric = load_metric("precision")
recall_metric = load_metric("recall")
f1_metric = load_metric("f1")

# Initialiser le tokenizer et le nom du modèle
model_name = "sentence-transformers/paraphrase-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Chargement des données
df_setfit_multilabel = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/df_setfit_multilabel.csv")

# Préparation des labels
df_setfit_multilabel['labels'] = df_setfit_multilabel['labels'].apply(lambda x: x.split(', '))
mlb = MultiLabelBinarizer()
labels_encoded = mlb.fit_transform(df_setfit_multilabel['labels'])
num_labels = len(mlb.classes_)

# les labels doit être du type float pour la BCE avec logits
labels_encoded = torch.tensor(labels_encoded).float()  # Convert labels to float tensors

# Fonction pour préparer les datasets
def prepare_dataset(df, labels_encoded):
    texts = df['translated_content'].tolist()
    encoded_inputs = tokenizer(texts, padding='max_length', truncation=True, max_length=4, return_tensors="pt")
    return Dataset.from_dict({
        'input_ids': encoded_inputs['input_ids'],
        'attention_mask': encoded_inputs['attention_mask'],
        'labels': labels_encoded
    })

# Séparation en ensembles de formation et de test
train_df, test_df, train_labels, test_labels = train_test_split(df_setfit_multilabel, labels_encoded, test_size=0.2, random_state=42)
train_dataset = prepare_dataset(train_df, train_labels)
eval_dataset = prepare_dataset(test_df, test_labels)

# Fonction pour calculer les métriques à partir des prédictions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = sigmoid(logits).round()  # Appliquer sigmoid défini manuellement et arrondir

    return {
        "accuracy": accuracy_score(labels, predictions),
        "precision": precision_score(labels, predictions, average='samples'),
        "recall": recall_score(labels, predictions, average='samples'),
        "f1": f1_score(labels, predictions, average='samples')
    }

# Fonction objectif pour l'optimisation avec Optuna et MLflow
def objective(trial):
    # Configuration de MLflow
    mlflow.set_tracking_uri('/content/drive/MyDrive/Colab Notebooks/mlruns_setfit_multilabel')
    mlflow.set_experiment("Setfit Multilabel Experiment")
    mlflow.start_run()

    # Suggestion des hyperparamètres
    learning_rate = trial.suggest_loguniform('learning_rate', 1e-5, 1e-3)
    batch_size = trial.suggest_categorical('batch_size', [2, 4, 8])
    num_train_epochs = trial.suggest_int('num_train_epochs', 1, 2)

    # Configuration des arguments de formation
    training_args = TrainingArguments(
        output_dir='/content/drive/MyDrive/Colab Notebooks/results_setfit_multilabel',
        evaluation_strategy="epoch",
        learning_rate=learning_rate,
        per_device_train_batch_size=batch_size,
        num_train_epochs=num_train_epochs,
        weight_decay=0.01,
    )

    # Initialisation du modèle et du trainer
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels, problem_type='multi_label_classification')
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=eval_dataset,
        tokenizer=tokenizer,
        compute_metrics=compute_metrics
    )

    # Entraînement du modèle
    trainer.train()
    eval_result = trainer.evaluate()

    # Logging des paramètres et des métriques dans MLflow
    mlflow.log_params({
        "learning_rate": learning_rate,
        "batch_size": batch_size,
        "num_train_epochs": num_train_epochs
    })
    mlflow.log_metrics({
        "accuracy": eval_result['eval_accuracy'],
        "precision": eval_result['eval_precision'],
        "recall": eval_result['eval_recall'],
        "f1": eval_result['eval_f1']
    })

    mlflow.end_run()

    return eval_result["eval_f1"]

# Exécution de l'étude Optuna
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=2)

print("Meilleurs hyperparamètres : ", study.best_trial.params)
```

```python
# Etape 2 : Re-former le modèle avec les meilleurs hyperparamètres

from transformers import TrainingArguments, Trainer

best_params = study.best_trial.params
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels, problem_type='multi_label_classification')

training_args = TrainingArguments(
    output_dir='/content/drive/MyDrive/Colab Notebooks/final_model_setfit_multilabel',
    evaluation_strategy="epoch",
    learning_rate=best_params['learning_rate'],
    per_device_train_batch_size=best_params['batch_size'],
    num_train_epochs=best_params['num_train_epochs'],
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

# Entraînement du modèle avec les meilleurs paramètres
trainer.train()

# Sauvegarde du modèle entraîné et du tokenizer
model.save_pretrained('/content/drive/MyDrive/Colab Notebooks/model_setfit_multilabel')
tokenizer.save_pretrained('/content/drive/MyDrive/Colab Notebooks/model_setfit_multilabel')
```

```python
# Etape 3 : Faire des Prédictions avec le Modèle Ajusté

from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Charger le tokenizer et le modèle depuis le répertoire de sauvegarde
tokenizer = AutoTokenizer.from_pretrained('/content/drive/MyDrive/Colab Notebooks/model_setfit_multilabel')
model = AutoModelForSequenceClassification.from_pretrained('/content/drive/MyDrive/Colab Notebooks/model_setfit_multilabel')
model.eval()

def predict(text, tokenizer, model):
    # Tokenisation et prédiction
    inputs = tokenizer(text, return_tensors="pt", padding='max_length', max_length=4, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)

    logits = outputs.logits
    predictions = logits.sigmoid().round().int()  # Utiliser sigmoid pour obtenir des prédictions binaires
    predicted_labels = [mlb.classes_[i] for i, value in enumerate(predictions[0]) if value == 1]

    return predicted_labels

# Exemple de test
test_text = "The staff was helpful but the product quality was poor."
print(predict(test_text, tokenizer, model))
```

```python
# Etape 1 : Optimisation des Hyperparamètres avec Optuna

import numpy as np
import pandas as pd
import optuna
import mlflow
from datasets import load_dataset, load_metric, Dataset, DatasetDict
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from torch import tensor

# Chargement des données
df_setfit_multilabel = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/df_setfit_multilabel.csv")

# Fonction pour étendre les données pour ABSA avec plusieurs aspects
def expand_rows(df):
    all_rows = []
    # Itérer sur chaque ligne du dataframe
    for _, row in df.iterrows():
        translated_content = row['translated_content']
        # Supprimer les crochets, splitter par virgule et nettoyer les espaces
        aspects = row['predicted_labels_setfit'].strip('[]').split(', ')
        # Pour chaque aspect, créer une nouvelle ligne avec toutes les informations de la ligne originale
        for aspect in aspects:
            # Créer une copie du dictionnaire de la ligne actuelle pour conserver toutes les colonnes
            new_row = row.to_dict()
            new_row['aspect'] = aspect.strip()  # Mettre à jour l'aspect
            all_rows.append(new_row)
    # Retourner un nouveau DataFrame basé sur la liste de tous les dictionnaires de lignes
    return pd.DataFrame(all_rows)

# Chargement des données
df = expand_rows(df_setfit_multilabel)


# Initialisation du tokenizer et du modèle
MODEL = "yangheng/deberta-v3-base-absa-v1.1"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
# num_labels=3 est pour indiquer trois catégories de sentiments (negative,neutral,positive)
model = AutoModelForSequenceClassification.from_pretrained(MODEL, num_labels=3)


def prepare_dataset(df):
    texts = df['translated_content'].tolist()
    aspects = df['aspect'].tolist()
    label_encoder = LabelEncoder()
    labels = label_encoder.fit_transform(aspects)  # Utiliser les étiquettes encodées pour le dataset
    merged_texts = [f"[CLS] {text} [SEP] {aspect} [SEP]" for text, aspect in zip(texts, aspects)]
    encoded_inputs = tokenizer(merged_texts, padding='max_length', truncation=True, max_length=4, return_tensors="pt")

    return Dataset.from_dict({
        'input_ids': encoded_inputs['input_ids'],
        'attention_mask': encoded_inputs['attention_mask'],
        'labels': tensor(labels)
    })

df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)

# Préparation des datasets
train_dataset = prepare_dataset(df_train)
eval_dataset = prepare_dataset(df_test)


# Métriques pour évaluer le modèle
accuracy = load_metric("accuracy")
precision = load_metric("precision")
recall = load_metric("recall")
f1 = load_metric("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy.compute(predictions=predictions, references=labels)["accuracy"],
        "precision": precision.compute(predictions=predictions, references=labels)["precision"],
        "recall": recall.compute(predictions=predictions, references=labels)["recall"],
        "f1": f1.compute(predictions=predictions, references=labels)["f1"]
    }

# Fonction objectif pour Optuna
def objective(trial):
    # Configuration de MLflow
    mlflow.set_tracking_uri("/content/drive/MyDrive/Colab Notebooks/mlruns_absa_multilabel")
    mlflow.set_experiment("ABSA Multilabel Experiment")
    mlflow.start_run()

    learning_rate = trial.suggest_loguniform('learning_rate', 1e-5, 1e-3)
    batch_size = trial.suggest_categorical("batch_size", [2, 4, 8])
    num_epochs = trial.suggest_int("num_epochs", 1, 2)

    training_args = TrainingArguments(
        output_dir="/content/drive/MyDrive/Colab Notebooks/results_absa_multilabel",
        evaluation_strategy="epoch",
        learning_rate=learning_rate,
        per_device_train_batch_size=batch_size,
        num_train_epochs=num_epochs,
        weight_decay=0.01
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=eval_dataset,
        compute_metrics=compute_metrics,
        tokenizer=tokenizer
    )

    trainer.train()
    eval_result = trainer.evaluate()

    # Enregistrement des résultats dans MLflow
    mlflow.log_params(trial.params)
    mlflow.log_metrics({
        "accuracy": eval_result["eval_accuracy"],
        "precision": eval_result["eval_precision"],
        "recall": eval_result["eval_recall"],
        "f1": eval_result["eval_f1"]
    })

    mlflow.end_run()
    # les classes sont desequilibré donc f1 va m'aider à minimiser à la fois les faux positifs et les faux négatifs,
    # cela permettre d'équilibrer au modèle la precision et le rappel
    return eval_result["eval_f1"]

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=2)

print("Meilleurs hyperparamètres : ", study.best_trial.params)
```

```python
# Etape 3 : Faire des predictions avec le meilleur modèle
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import torch

input_dir = "/content/drive/MyDrive/Colab Notebooks/absa_multilabel_model"

# Charger le tokenizer et le modèle depuis le répertoire de sauvegarde
tokenizer = AutoTokenizer.from_pretrained(input_dir)
model = AutoModelForSequenceClassification.from_pretrained(input_dir, num_labels=3)
model.eval

def predict(text, aspect, tokenizer, model):

    # Préparation de l'entrée
    inputs = tokenizer(f"[CLS] {text} [SEP] {aspect} [SEP]", return_tensors="pt", padding=True, truncation=True, max_length=512)

    # Prédiction
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        probabilities = torch.softmax(logits, dim=-1)
        predicted_class = torch.argmax(probabilities, dim=-1)
        predicted_label = ['negative', 'neutral', 'positive'][predicted_class.item()]
        predicted_score = probabilities[0, predicted_class.item()].item()

    return predicted_label, predicted_score

# Exemple de test
text = "The food was great but the service was terrible"
aspect = "service"
predicted_label, predicted_score = predict(text, aspect, tokenizer, model)
print(f"Predicted label: {predicted_label}, Score: {predicted_score}")
```
